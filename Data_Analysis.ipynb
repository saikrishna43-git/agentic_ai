```python

import plotly.express as px

# Static pie chart for Product using matplotlib
product_counts = df['Product'].value_counts()
plt.figure(figsize=(4,4))
plt.pie(product_counts, labels=product_counts.index, autopct='%1.1f%%', startangle=90, wedgeprops={'edgecolor': 'black'})
plt.title('Distribution of Products (Static)')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()

# Interactive pie chart for Product using plotly
fig = px.pie(df, names='Product', title='Interactive Distribution of Products')
fig.show()
```
